```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
surveys = pd.read_csv('/content/Short Patient Survey.csv')
```

```python
surveys
```

```python
tipos_de_datos = surveys.dtypes
print(tipos_de_datos)
```

```python
proyecto = surveys[['Respondent ID', 'End Date', 'language', 'Do you have any other comments or concerns?']]
proyecto['End Date'] = pd.to_datetime(proyecto['End Date'])
proyecto['End Date'] = proyecto['End Date'].dt.date
```

```python
proyecto
```

```python
surveys_english = proyecto[proyecto['language'] == 'en']
surveys_spanish = proyecto[proyecto['language'] == 'es_US']
```

```python
surveys_english
```

```python
surveys_spanish
```

```python
# Contar nulos en cada base de datos
nulos_english = surveys_english.isnull().sum()
nulos_spanish = surveys_spanish.isnull().sum()

print("Nulos en encuestas en inglés:\n", nulos_english)
print("\nNulos en encuestas en español:\n", nulos_spanish)
```

```python
surveys_spanish.dropna(inplace=True)
surveys_spanish
```

```python
surveys_english.dropna(inplace=True)
surveys_english
```

```python
import matplotlib.pyplot as plt

# Crear una gráfica temporal para las encuestas en inglés
plt.figure(figsize=(12, 6))
plt.hist(surveys_english['End Date'], bins=30, color='#00bfb2ff', alpha=0.7, label='English Surveys')
plt.hist(surveys_spanish['End Date'], bins=30, color='#009ace', alpha=0.7, label='Spanish Surveys')
plt.xlabel('Date')
plt.ylabel('Number of Surveys')
plt.title('Temporal Distribution of Surveys')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

estadisticas_ingles = surveys_english['word_count'].describe()
estadisticas_espanol = surveys_spanish['word_count'].describe()

df_descriptivos = pd.DataFrame({
    'Estadística': ['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'],
    'Inglés': estadisticas_ingles.values,
    'Español': estadisticas_espanol.values
})

print("Resumen Descriptivo de la Longitud de Comentarios:\n", df_descriptivos)

percentiles_english = surveys_english['word_count'].quantile([0.25, 0.5, 0.75])
percentiles_spanish = surveys_spanish['word_count'].quantile([0.25, 0.5, 0.75])

df_percentiles = pd.DataFrame({
    'Percentil': ['25%', '50%', '75%'],
    'Inglés': percentiles_english.values,
    'Español': percentiles_spanish.values
})

print("\nTabla de Percentiles:\n", df_percentiles)

```
